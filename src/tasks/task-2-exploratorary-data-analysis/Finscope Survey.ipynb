### Checking the ratio of bank account holders to non bank account holders

```python
sns.countplot(data=df_train, x='bank_account')
```

```python
df_train['bank_account'].describe()
```

The above plot confirms that the number of individuals with bank accounts is less as compared to those with bank accounts. From the view we can say more than 80% dont have bank accounts 

```python
sns.countplot(data=df_train, x='bank_account', hue='country')
```

In terms of owning a bank account and in terms of the country an individual comes from, we get to see that kenya has more bank account holders while uganda has the least. also Rwanda has the largest number of non account holders with uganda having the least too. for uganda we might consider the fact that data from this country was less hence the lower number

```python
sns.countplot(data=df_train, x='bank_account', hue='location_type')
```

while considering an individuals location, people living in the rural location most dont have bank accounts. but if we compare the number of those living in urban areas and rural areas for account holders we have almost an equal number. 

```python
sns.countplot(data=df_train, x='bank_account', hue='gender_of_respondent')
```

More females prefer not having a bank account as compared to males 

```python
sns.countplot(data=df_train, x='bank_account', hue='education_level')
```

When looking at education level, we get to see that individuals with no formal education are the least contribute less to the number of bank account holders followess by those who go to vocational training. 

for non bank account holders we get to see that those who reach primary education level are the most
